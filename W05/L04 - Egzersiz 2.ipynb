```python
# Soru 3
# country sutununun 'Australia' ve 'New Zealand' degerinde olan ve points sutununu 95'e esit veya yuksek olan degerleri indexleyin, burada isin() metodunu kullanmaniz gerekiyor

df[ (df.points > 98 ) & ( (df.country == 'Australia') | (df.country == 'New Zealand') ) ]
# ya da 

# df.country.isin(['Australia', 'New Zealand'])  # seklinde de kullanilabiliyor

df[ (df.points > 98 ) & df.country.isin(['Australia', 'New Zealand']) ]

# hoca iloc ile de yazilabiliyor
```

```python
# soru 4 icinde italy ve france olmayanlari indexleyelim

df[ (df.points > 98 ) & (df.country.isin(['Italy', 'France']) == False) ].head(5)
# ya da
df[ (df.points > 98 ) & (~df.country.isin(['Italy', 'France']) ) ].head(5)   # tilda isareti
```
