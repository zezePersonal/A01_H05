```python
# median, mod, mean
# istatistik yontemleri olarak en cok kullanilan veri doldurma yontemleri

# gercekten doldurmali miyim, ki bu ornekte 3300 dolar sapittiriyor, median daha mantikli
# kaldirilmasi da diger secenek, bu ornekte bence cok rahat kaldirilir
```

```python
# hoca dolduralim dedi

median_price = df.price.median()
df['price'].fillna(median_price, inplace=True)  # NaN'lari median ile doldur
```

```python
# Hoca ayni mantikta numeric degiskenleri object olarak degistirdi
```

```python
object_cols = [
    column
    for column in df.columns
    if (df[column].dtype == 'object') and any( df[column].isnull() )
            
]
# null degerleri olan fieldler 

object_cols
```

```python
for col in df[object_cols]:
    df[col].fillna(df[col].mode(), inplace=True)
```
